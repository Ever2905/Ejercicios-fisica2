# ejercicio 3

Ejercicio:3  ¿que fuerza debemos ejercer en el embolo pequeño de una prensa hidraulica Si 
queremos levantar una masa de 20000kg situada en el embolo grande.el radio de ambos
embolos es de 2m y 4m, respectivamente.?

Formula:  $p=\frac{F_1}{A_1}=\frac{F_2}{A_2}$

### Datos:
    
   - $m=20000kg$
   - $r_1=0.2m$
   - $r_2=4m$

### Incognita:

$F_1=\space ?$

### Procedimiento:
Queremos calcular la fuerza que hay que aplicar en el embolo pequeño para levantar 20000 kg
en el embolo grande

Formula:
$\frac{F_1}{A_1}=\frac{F_2}{A_2}$

 

 Queremos obtenerla ${F_1}$.necesitamos despejar ${F_1}$ de la ecuacion
             
 $A_1\frac{F_1}{A_1}=A_1\frac{F_2}{A_2}$

$F_1=A_1\frac{F_2}{A_2}$

Nececitamos la ${F_2}$.se puede obtener calculando el peso,el cual se obtiene multiplicando masa 
por gravedad.

$F_2 = m*g$

$F_2 = 20000kg*9.8\frac{m}{s^2}$

$F_2 = 196000 N $

Necesitamos el ${A_1}$.se puede obtener con la formula $ {A} =\pi*r^2$ 

$ {A_1} =\pi*r_1$ 


$ {A_1} =\pi*{0.2m}^2$

$ {A_1} =\pi*{0.2m}^2$

$ {A_1} =\pi*0.04m^2$
   
 $ {A_1} =0.13m^2$ 


Necesitamos el ${A_2}$.se puede obtener con la formula $ {A} =\pi*r^2$ 

$ {A_2} =\pi*r_2^2$ 

$ {A_2} =\pi*{4m}^2$

$ {A_2} =\pi*{4m}^2$

$ {A_2} =\pi*16m^2$

$ {A_2} =50.27m^2$

Conlos datos anteriores ya podemos calcular $F1$

$F1= 0.13m^2\frac{196000N} {50.27m^2} $

### Resultado final:

$F1= 506.86N$
